```python
import pandas as pd
```

```python
rogames = pd.read_csv("roblox_games_data.csv")
```

```python
#Removing duplicates and keeping the ones with the latest update
rogames['Last Updated'] = pd.to_datetime(rogames['Last Updated'])
rogames = rogames.sort_values(by = 'Last Updated', ascending = False)
rogames = rogames.drop_duplicates(subset = 'gameID', keep = 'first') # Checks the gameID and keeps first occurence of duplicate
rogames = rogames.reset_index(drop = True)
rogames.fillna('', inplace = True)
```

```python
import re

def clean_title(title):
    title = re.sub(r"[\[\(].*?[\]\)]", "", title) # Removing text inside [] or () brackets
    title = re.sub(r"(?<=[a-z])(?=[A-Z])", " ", title) #Adds spaces between lowercase and uppercase words but not all uppercase words
    title = re.sub(r"by@?.*$", "", title, flags=re.IGNORECASE) # Removing the creator of the game from the title
    return title
```

```python
rogames["Cleaned Title"] = rogames["Title"].apply(clean_title)
```

```python
rogames['Combined'] = rogames['Cleaned Title'] + ' ' + rogames['Genre'] + ' ' + rogames['Description']
```

```python
rogames
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(ngram_range=(1,2)) # Will also look at groups of 2 words that are consecutive
tfidf = vectorizer.fit_transform(rogames["Cleaned Title"])
tfidf_combined = vectorizer.fit_transform(rogames["Combined"])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def search(game_title):
    game_title = clean_title(game_title)
    query_vec = vectorizer.transform([game_title])
    similarity = cosine_similarity(query_vec, tfidf).flatten() # Compares vectorizer to all of the cleaned titles in the dataset and return how similar it is
    indices = np.argpartition(similarity, -10)[-10:] # Finds 5 titles with the greatest similarity to game_title
    results = rogames.iloc[indices][::-1] # Reverses results and puts the most similar game from last to first
    return results[['Cleaned Title', 'Genre', 'Description', 'gameID']]

def search_genre(genre):
    genre = genre.lower()
    filtered = rogames[rogames['Genre'].str.lower() == genre]
    results = filtered.sort_values(by='Favorites', ascending=False) # Puts the most favorited games first
    return results[['Cleaned Title', 'Genre', 'Description', 'gameID']]

def search_combined(user_input):
    game_title = clean_title(user_input)
    query_vec = vectorizer.transform([game_title])
    similarity = cosine_similarity(query_vec, tfidf_combined).flatten()
    indices = np.argpartition(similarity, -10)[-10:]
    results = rogames.iloc[indices][::-1]
    return results[['Cleaned Title', 'Genre', 'Description', 'gameID']]
```

```python
import ipywidgets as widgets
from IPython.display import display

game_input = widgets.Text(
    description = "Title Search:",
    disabled = False
)

genre_input = widgets.Text(
    description = "Genre Search:",
    disabled = False
)

combined_input = widgets.Text(
    description = "Smart Search:",
    disabled = False
)
    
game_list = widgets.Output()
genre_list = widgets.Output()
combined_list = widgets.Output()

def on_type(data):
    with game_list:
        game_list.clear_output()
        title = data["new"]
        if len(title) > 0:
            display(search(title))
            
def on_genre(data):
    with genre_list:
        genre_list.clear_output()
        genre = data["new"]
        if len(genre) > 0:
            display(search_genre(genre))

def on_combined(data):
    with combined_list:
        combined_list.clear_output()
        user_input = data["new"]
        if len(user_input) > 0:
            display(search_combined(user_input))

game_input.observe(on_type, names='value')
genre_input.observe(on_genre, names='value')
combined_input.observe(on_combined, names='value')

display(game_input, game_list)
display(genre_input, genre_list)
display(combined_input, combined_list)
```
